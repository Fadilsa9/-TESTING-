### Зачем нужны функции?

С ростом программ, которые мы пишем, возрастает необходимость в их максимальном упрощении и облегчении их поддержки и сопровождения.
Одним из способов добиться этого является дробление программ на отдельные блоки, именуемые **функциями**.

Функции в программах служат сразу нескольким целям:
1. Они позволяют написать код один раз и впоследствии обращаться к нему из разных мест.
2. Позволяют тестировать различные решения.
3. Позволяют "скрыть" детали реализации программного кода.

С помощью функций программист имеет возможность написать необходимое количество строк кода, спрятать их в единый блок - функцию, и отложить для использования в будущем. К этому коду в дальнейшем можно будет обращаться из других блоков и программ.

Блок функции со всеми входящими в него инструкциями именуется **определением функции**, а обращение к  нему извне – **вызовом функции**. После вызова и  выполнения функции управление возвращается в основную программу, и  она продолжается с первой после вызова функции инструкции.

Программы, которые мы писали до этого, содержат вызовы функций вроде *print*, *input*, *int* и *float*. Все эти функции были заранее написаны
разработчиками языка Python и могут быть вызваны в любой программе на этом языке.

Определение функции начинается с  ключевого слова **def**, следом за которым идет имя функции, открывающая и закрывающая скобки и знак двоеточия. Далее следует блок с набором инструкций и  отступами (как в  случае с циклами и условиями), которые и составляют тело функции.

Функция завершается тогда, когда исполнение кода доходит до конца блока с отступами. Т.е. когда следующая инструкция программы уже будет на одном уровне (одинаковым отступом) с ключевым словом *def*.

````python
def myFunction():
    print("**********")
    print("Это моя первая функция!")
    print("**********")
````

Теперь когда функция объявлена, ее необходимо вызвать:

````python
myFunction()
````

Как видно выше, объявление функции не приводит к ее выполнению. Объявление функции откладывает на будущее заключенные в ней строки кода, ассоциируя их с выбранным именем функции.

Чтобы вызвать функцию в программе, мы обратились к ней по имени со скобками. Можно вызвать любое количество функций подряд:

````python
myFunction()
myFunction()
myFunction()
````

````python
# https://monica.im/home/chat/Monica/monica?convId=conv%3A6ebd0b5e-f3ba-45f9-8a5f-98135772af05

## Рисуем прямоугольник из звездочек, заполненный пробелами

def drawRectangle(width, height):
    if width > 2 and height > 2:
        # Рисуем верхнюю линию прямоугольника
        print(" * " * width)
        # Рисуем стороны прямоугольника
        for i in range(height - 2):
            print(" * " + "   " * (width-2) + " * ")
        # Рисуем нижнюю линию прямоугольника
        print(" * " * width)

    else:
        print("Ошибка: ширина или высота прямоугольника слишком малы.")

drawRectangle(6, 8)
````

Переменные, которые мы объявили внутри функции, называются локальными. Это означает, что область действия (видимость) этих переменных ограничивается
данной функцией, и к ним можно обратиться только в теле этой функции.
После завершения работы функции локальные переменные очищаются из памяти, а значит, и обратиться к ним больше не получится. К локальным переменным относят не только переменные объявленные в самой функции, но и переменные параметров вроде *width* и *height*. Ко всем этим переменным можно обращаться только внутри тела функции. Переменные, создаваемые и инициализируемые внутри функции при помощи оператора присваивания, также классифицируются как локальные.


### Как из функции вернуть результат выполнения?

Многие функции пишутся для того, чтобы провести какие-нибудь сложные вычисления со входными параметрами и вернуть в основную программу результат вычислений.

Например, функция *sqrt* из модуля *math* вычисляет квадратный корень из переданного в  нее аргумента и  возвращает его для дальнейшего использования. Так же точно функция *input* считывает ввод с  клавиатуры и возвращает введенное значение для дальнейшей его обработки. Многие функции, которые вы будете писать, тоже будут возвращать значения.

Для возврата значения из функции существует ключевое слово **return**, после которого идет само значение. Дойдя до инструкции с *return*, функция мгновенно прекращает свое выполнение, и управление передается инструкции, следующей сразу за вызовом функции.

Рассмотрим пример:


````python
def number5():
    return 5

num = number5()
print(num)
````

В примере выше значение, которое вернула функция, было присвоено переменной *num*.

Функции могут использоваться в выражениях справа от оператора присваивания, могут быть частью условного выражения *if* или цикла *while*, а также подаваться на вход других функций, таких как *print* или *range*.

В функциях, не предназначенных для возвращения результата, не обязательно использовать ключевое слово return, поскольку возврат в основную программу автоматически произойдет после выполнения последней инструкции в теле функции. Но программист вправе использовать слово *return* без следующего за ним значения для преждевременного выхода из функции. Кроме того, в любой функции, вне зависимости от того, возвращает она значение или нет, ключевое слово return может появляться множество раз. Выход из подобной функции произойдет по достижении первого из этих слов:



````python
# https://monica.im/home/chat/Monica/monica?convId=conv%3Ae858a85e-f856-44c7-be21-4cc7f6cea05b

# Функция проверяет, является ли число простым. В случае, если число простое - вернет True, иначе False

def isPrime(n):
    if n <= 1:
        return False

    # Проверяем все числа от двух до n, не включая его, на деление n на них без остатка
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

value = int(input("Введите целое число: "))
if isPrime(value):
    print(value, "– простое число.")
else:
    print(value, "не является простым числом.")
````

### Функция с переменным числом аргументов: `*args` и `**kwargs`

Не всегда заранее известно, сколько аргументов будет передано функции. В таких случаях в Python на помощь приходят `*args` и `**kwargs`, позволяющие передавать переменное количество аргументов.

Предположим, что нам надо написать функцию для суммирования всех переданных ей чисел. Например, если надо сложить только 3 числа, то здесь все просто, передаем 3 аргумента:

````python
def sum3(x, y, z):
    print("сумма:", x + y + z)

sum3(10, 12, 13)
````

Но если параметров 5, 10 или неизвестное число, то как быть?

В Python можно передать переменное количество аргументов двумя способами:

* `*args` - для неименованных аргументов;
* `**kwargs` - для именованных аргументов.

Мы используем `*args` и `**kwargs` в качестве аргумента, когда заранее не известно, сколько значений мы хотим передать функции.

##### ***args**

Если поставить `*` перед именем, это имя будет принимать не один аргумент, а несколько. Аргументы передаются как кортеж и доступны внутри функции под тем же именем, что и имя параметра, только без `*`:

````python
def sumN(*nums):
    sum = 0
    for n in nums: # Перебор по кортежу
        sum += n

    print("Сумма:", sum)

sumN(1, 9) # Передаем 2 аргумента
sumN(10, 2, 2, 11) # Передаем 4 аргумента
sumN(8, 12, 6, 23, 67, 2, 1, 4)  # Передаем 8 аргументов
````

Однако, программисты чаще в коде используют `*args` (а не как мы с вами `*nums`), это своего рода неформальная договоренность, которая помогает быстрее находить ошибки в "чужом коде". Ведь `*` - всегда можно перепутать со знаком умножения, а тут сразу понятно. Серьезные программы пишут серьезные программисты, которые ценят свое время, причем пишут большими группами, поэтому они между собой договариваются о всяких мелочах, которые позволяют экономить время на исправления в "чужом коде". ##### ****kwargs**

По аналогии с `*args` мы используем `**kwargs` для передачи переменного количества именованных аргументов. Если поставить `**` перед именем, то это имя будет принимать любое количество именованных аргументов. В данном случае передаваться уже будет не кортеж, а словарь. А как мы помним, словарь имеет именованные ключи. Поэтому к параметрам внутри функции мы сможем обращаться по имени:

Пример: Код выполняет задачу **вывода информации о пользователе** на основе переданных именованных аргументов (например, имя, фамилия, возраст). Он проверяет наличие ключа `'firstname'`, выводит его значение, а затем перечисляет все переданные параметры и их значения.

````python
# https://monica.im/home/chat/Monica/monica?convId=conv%3A4b31e682-c862-47d9-8447-b631e5ffee74

def intro(**kwargs): # Используем kwargs, вместо произвольной переменной, заботимся о тех, кто будет читать ваш код
    print()
    print("Тип данных kwargs: ", type(kwargs)) # Доказательство, что передается словарь
    print("Список кортежей:", kwargs.items()) # Функция items() для перебора (ключ, значение)
    if 'firstname' in kwargs: # Проверяем есть ли среди параметров имя?
        print("Имя:", kwargs['firstname']) # Выводим значение по ключу

    for key, value in kwargs.items(): # Перебираем все параметры
        print("{} is {}".format(key, value))

intro(firstname="Ivan", lastname="Ivanov", age=22)
intro(firstname="Igor", lastname="Petrov", email="igor@nomail.com", country="Russia")
````

### Lambda-функция и функциональный стиль программирования

**Lambda-функция** (или анонимная функция) - это способ создания простых функций в Python без необходимости использования стандартного синтаксиса определения функции с помощью ключевого слова **def**. Т.е. это функция без объявления, создается "на лету" в том месте, где ее необходимо использовать. Lambda-функции не требуют имени и определяются в точке использования. Они ограничены одним выражением и не имеют блока кода. Чаще всего они используются для выполнения простых операций, таких как преобразование данных или фильтрация:

```python
lambda arguments: expression

```

`arguments` - список аргументов функции.

`expression` - это выражение, которое выполняется и возвращается как результат работы функции.

Напишем функцию возведения в степень:

````python
def pow(a, b): # Функция возводит a в степень b
    return a ** b

print(pow(2, 8))
````

А теперь представим эту же функцию в виде lambda-функции:

````python
# Lambda-функция для возведения в степень
pow = lambda a, b: a ** b

print(pow(2,8))
````

Вся мощь lambda-функций проявляется при функциональном стиле программирования в Python.

В функциональном программировании вычисления выполняются путем объединения функций, которые принимают аргументы и возвращают конкретное значение (или значения). Эти функции не изменяют свои входные аргументы и не изменяют состояние программы. Они просто предоставляют результат данного вычисления. Такие функции обычно называются чистыми функциями (pure functions).

Функциональное программирование обычно использует списки, массивы и другие итерационные объекты для представления данных вместе с набором функций, которые работают с этими данными и преобразовывают их.

Когда дело доходит до обработки данных в функциональном стиле, обычно используются как минимум три метода:

1. **Сопоставление, отображение (Mapping)**.
2. **Фильтрация (Filtering)**.
3. **Сокращение (Reducing)**.

Рассмотрим реализацию этих методов в Python.

#### Функция map()

Довольно часто возникает задача, когда необходимо применить специальную функцию для всех элементов в последовательности. В функциональном программировании она называется *сопоставлением* или *отображением*.

Встроенная в Python функция *map* позволяет обрабатывать одну или несколько последовательностей с использованием заданной функции. Рассмотрим функцию map:

```python
map(функция, последовательности)
```
В данном виде `функция` – это ссылка на стандартную функцию либо lambda-функция, а `последовательности` – это одна или несколько отделенных запятыми итерируемых последовательностей, т.е. списки, кортежи, диапазоны или строковые данные.


````python
seq = (1, 2, 3, 4, 5, 6, 7, 8, 9) # Обратите внимание, что мы передаем последовательности разной длины
seq2 = (5, 6, 7, 8, 9, 0) # Результат будет определяться меньшей по длине последовательностью
result = map(lambda x, y: x * y, seq, seq2) # Применяем lambda-функции ко всем элементам последовательности
print(result) # Результат ленивых вычислений
print(list(result)) # Преобразуем в список
````

В приведенном выше примере, с помощью функции map(), мы применяем к нашим последовательностям seq и seq2, lambda-функцию.  Обратите внимание, что функция map возвращает объект-последовательность map (map object). Особенность объекта-последовательности map состоит в том, что он может предоставлять свои элементы, только когда они потребуются, используя ленивые вычисления.

Ленивые вычисления – это стратегия вычисления, согласно которой вычисления следует откладывать до тех пор, пока не понадобится их результат.

Программистам часто приходится обрабатывать последовательности, состоящие из десятков тысяч и даже миллионов элементов. Хранить их в оперативной памяти, когда в определенный момент нужен всего один элемент, не имеет никакого смысла. Ленивые вычисления позволяют генерировать ленивые последовательности, которые при обращении к ним предоставляют следующий элемент последовательности. Чтобы показать ленивую последовательность, в данном случае результат работы примера, необходимо эту последовательность вычислить. Для этого преобразуем наш объект-последовательность в список с помощью функции list().

#### Функция filter()

Функция **filter()** обычно используется для фильтрации данных.

```python
filter(предикативная_функция, последовательность)
```

В данном виде `предикативная_функция` – это ссылка на стандартную функцию либо lambda-функция, которая возвращает True либо False, и `последовательность` – это итерируемая последовательность, т.е. список, кортеж, диапазон или строковые данные.

Приведем пример фильтрации только четных чисел:



````python
# lambda-функция вернет True если число четное
is_even = lambda x: x % 2 == 0

# Применим lambda-функцию в качестве предиктивной функции фильтрации к последовательности
seq = (1, 2, 3, 4, 5, 6, 7, 8, 9)
filtered = filter(is_even, seq)
print(list(filtered))
````

#### Функция reduce()

Функция **reduce()** используется с целью обработать список значений таким образом, чтобы свести последовательность к единственному результату. Функция reduce находится в модуле functools стандартной библиотеки и ее необходимо импортировать:
```python
from functools import reduce
```

Рассмотрим функцию reduce:

```python
reduce(функция, последовательность, инициализатор)
```

В данном представлении:

 `функция` – это ссылка на редуцирующую функцию, ею может быть стандартная функция либо lambda-функция,

`последовательность` – это итерируемая последовательность, т.е. список, кортеж, диапазон или строковые данные,

`инициализатор` – это параметрическая переменная, которая получает начальное значение для накопителя.

Начальным значением может быть значение любого базового типа данных, а также быть списком, кортежем и т.д. Начальное значение инициирует накапливающую переменную, которая прежде чем она будет возвращена, будет обновляться редуцирующей функцией по каждому элементу в списке.

Переданная при вызове функция вызывается в цикле для каждого элемента последовательности. Например, функция reduce может применяться для суммирования числовых значений в списке. Например, вот так:

````python
from functools import reduce
seq = (1, 2, 3, 4, 5, 6, 7, 8, 9)

summed_numbers = reduce(lambda a, b: a + b, seq)
summed_numbers
````

Рассмотрим более сложный пример. Пусть у нас есть seq - последовательность предложений, и требуется подсчитать общее количество слов в этих предложениях:

````python
seq = [
    "Унылая пора! Очей очарованье!",
    "Приятна мне твоя прощальная краса —",
    "Люблю я пышное природы увяданье,",
    "В багрец и в золото одетые леса,",
    "В их сенях ветра шум и свежее дыханье,",
    "И мглой волнистою покрыты небеса,",
    "И редкий солнца луч, и первые морозы,",
    "И отдаленные седой зимы угрозы."]

number_of_words = reduce(lambda aсс, seq: aсс + len(seq.split()), seq, 0)
print("Число слов в последовательности:", number_of_words)
````

В примере выше, в lambda-функции строковый метод split разбивает предложение на список слов, функция len подсчитывает количество элементов в получившемся списке и прибавляет его в накапливающую переменную.

Светофор, автомобиль, пешеход, кошка, собака, самолет - это все объекты.

Рассмотрим более подробно такой объект, как кошка.
Абсолютно все кошки обладают некоторыми характерными свойствами, или параметрами. К ним можно отнести наличие хвоста, четырех лап, шерсти, усов и т. п. Но это еще не все. Помимо определенных внешних параметров, кошка может выполнять свойственные ей действия: мурлыкать, шипеть, царапаться, кусаться.

Теперь возьмем такой объект, как автомобиль. Абсолютно все автомобили имеют
кузов, колеса, фары, двигатель, тормоза и т. п. Помимо этих внешних параметров,
автомобиль может выполнять свойственные ему действия: двигаться, тормозить, издавать звуковой сигнал, выпускать выхлопные газы и т. п.

Мы представили некие общие свойства всех кошек и всех автомобилей в целом, т. е. описали два класса: класс кошка и класс автомобиль. Подобное описание характерных свойств и действий какого-либо объекта на различных языках программирования и называется **классом**.

**Класс** — просто набор переменных и функций, которые описывают какой-то объект.

Если все кошки (как класс) имеют общую характеристику — наличие шерсти, то
каждая конкретная кошка (как объект) будет иметь свойственную только ей длину
шерсти и ее окрас. То же самое касается и автомобиля: все автомобили (как класс)
имеют кузов, но каждый конкретный автомобиль (как объект) имеет свою форму
кузова, свой цвет. Все автомобили имеют колеса, но на каждом конкретном автомобиле стоят шины определенного размера и завода-изготовителя. Исходя из сказанного, очень важно понимать разницу между классом и конкретным объектом
этого класса. Класс — это некая схема, которая описывает объект в целом. Объект
класса — это, если можно так сказать, материальное воплощение некого конкретного элемента из этого класса. Класс кошка — это описание ее свойств и действий, он всегда только один. А объектов-кошек может быть великое множество. По аналогии класс Автомобиль — это описание свойств и действий автомобиля, такой класс всегда только один. А объектов-автомобилей может очень много.


Для создания класса надо написать ключевое слово *class* и затем указать имя этого класса.

Создадим класс кошек с именем *Cat*:


````python
class Cat:
    pass # нулевой оператор, используемый в качестве заполнителя для будущего кода
````


Далее нужно указать действия, которые будет способен совершать этот класс
(в нашем случае - возможные действия кошки). Такие действия реализуются в виде функций, которые описываются внутри класса. Функции, прописанные внутри класса, называются методами. Словесно мы можем описать следующие методы кошки (действия, которые она может совершать): мурлыкать (purr), шипеть (hiss), царапаться (scrabble). Теперь реализуем эти методы в классе кошки *Cat* на
языке Python:

````python
# Класс кошки
class Cat:
    # Мурлыкать
    def purr(self):
        print("Myppp!")

    # Шипеть
    def hiss(self):
        print("Шшшш!")

    # Царапаться
    def scrabble(self):
        print("Цап-царап!")
````

Как видно из приведенного примера, каждая из трех описанных функций имеет
единственный параметр: *self*. В методах класса первый параметр функции по соглашению именуют *self*, и он представляет собой ссылку на сам объект этого класса. Помещать такой параметр нужно в каждую функцию, чтобы иметь возможность вызвать ее на текущем объекте. Таким образом, *self* заменяет идентификатор объекта. Все наши кошки (объекты cat) способны мурлыкать (имеют метод purr).
Теперь мы хотим, чтобы замурлыкал наш конкретный объект-кошка, созданный на основе класса Cat. Как это сделать?
Допустим, мы в Python напишем следующую команду:
```
Cat.purr()
```
Если мы запустим на выполнение эту команду, то будут мурлыкать сразу все коты

```
то будет мурлыкать только та особь, на которую укажет параметр self, т. е. только наша кошка.


Как видите, обязательный для любого метода параметр *self* позволяет нам обращаться к методам и переменным самого класса! Без этого аргумента выполнить
подобные действия мы бы не смогли.

Кроме того, что кошки могут шипеть, мурлыкать и царапаться, они имеют еще и

В абсолютно любом классе нужно определять функцию: `__init__`. Она вызывается всегда, когда мы создаем реальный объект на основе нашего класса (обратите внимание, что здесь используются символы двойного подчеркивания).

В следующем примере, зададим нашим кошкам наименование класса *Name_Сlass*, и три свойства: окрас шерсти — *wool_color*, цвет глаз — *eyes_color*, кличку — *name*:

В приведенном здесь методе __init__, мы задаем переменные, в которых будут храниться свойства нашей кошки.
Как мы это делаем? Обязательно используем параметр *self*, чтобы при создании конкретного объекта, кошки, сразу (по умолчанию) задать ей три нужных свойства. Затем мы определяем в этом методе три переменные, отвечающие за окрас шерсти, цвет глаз и кличку.

```

Зачем они нужны?

Достаточно
много программистов создают приложения с полезными функциями и классами.
Другие программисты могут подключать эти сторонние модули и использовать все
имеющиеся в них функции и классы, тем самым упрощая себе работу.
Например, вам не нужно тратить время и писать свой программный код для работы
с матрицами. Достаточно подключить модуль *NumPy* и использовать его функции
и классы. Модуль *NumPy* позволяет быстро и удобно работать с матрицами и многомерными массивами. А, например, модуль *math* предоставляет множество методов для работы с числами: синусы, косинусы, переводы градусов в радианы и пр.


Еще одна из причин использовать модули - это повторное использование написанного вами ранее кода. Многие программисты, даже самые опытные, предпочитают просто копировать код функций из старых файлов и вставлять в новые, но это не лучший подход. Копирование функции приводит к ее полному дублированию, и если вам вдруг придется внести изменения в работу функции, вы вынуждены будете делать это во всех файлах.

Гораздо лучше просто импортировать функцию из старого файла в новый так же точно, как мы импортировали функции из встроенных в Python модулей, – при помощи ключевого слова *import* с последующим указанием имени файла (без расширения .py), в котором находятся нужные функции. Это позволит обращаться к  нужным функциям, но при этом будет также запущена программа из старого файла. В некоторых ситуациях это допустимо, но не всегда.

Как же решить эту проблему?

В подобном случае обычно создают функция с именем *main*, которая содержит необходимые выражения для решения задачи. В конце кода ставится вызов функции *main*. И наконец, вставляется условная конструкция, приведенная ниже, предотвращающая запуск функции main при выполнении импорта из другого файла.
```
if __name__ == "__main__":
    main()
```

Такую структуру файла стоит использовать всякий раз, когда вы знаете, что написанные в нем функции в будущем могут пригодиться в других программах.
