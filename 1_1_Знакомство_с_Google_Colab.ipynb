Google Colaboratory - это бесплатная платформа для интерактивных блокнотов Jupyter. В ней можно запускать блокноты как на языке Python, так и R. А самое главное, что все блокноты можно использовать совместно.

Jupyter, а следовательно и Google Colaboratory - это программное обеспечение, которое объединяет в едином редактируемом документе (блокноте):

### Ячейки Colab (блокнота)

В блокноте можно создавать ячейки двух видов: **текстовые ячейки** и **ячейки с кодом**.

**Текстовые ячейки** предназначены, как следует из названия, для создания текстовых блоков, вставки видео и изображений. До текущего момента в блокноте использовались только текстовые ячейки. В текстовых ячейках используется специальный синтаксис, который называется **Markdown**. Он широко применяется среди разработчиков для описания своих проектов (документирования) в репозиториях (их также называют Репы).
Самая популярная репа среди разработчиков - это **GitHub**.

**GitHub** — веб-сервис, который основан на системе **Git**. Это такая социальная сеть для разработчиков, которая помогает удобно вести коллективную разработку IT-проектов. Здесь можно публиковать и редактировать свой код, комментировать чужие наработки, следить за новостями других пользователей.

**Git** - это распределённая система контроля версий, которую создал Линус Торвальд для управления разработкой ядра Linux. Как вы, наверное, уже догадались, именно он является разработчиком операционной системы  Linux. Как он в шутку выражался, что Linux он назвал в честь себя, так же как и git. **Ячейки с кодом** - это ячейки, в которых вы пишите свою программу и выполняете код. Только делаете это последовательно, ячейку за ячейкой. Это довольно удобный подход. Если у вас возникла ошибка, то не обязательно выполнять весь код заново, достаточно перезапустить эту ячейку. И все последующие.

Для того чтобы научиться работать с ячейками решим простую задачу.

**Задача. С какой скоростью надо кинуть камень, чтобы он стал спутником Земли?**

image.png

```python
# с минимальной орбитальной скоростью, которая составляет примерно 7,9 км/с.

# Подставляя численные значения для орбиты, расположенной вблизи поверхности Земли
# (h ≈ 0, M = 5,97·10^24 кг, R0 = 6 371 000 м (радиус указывается в метрах), G=6.67·10^-11 м³·кг⁻¹·с⁻²), получаем
```

Для того чтобы создать текстовую ячейку и вставить в неё описание нажмите на кнопку **"+ Текст"** (**"+ Тт"** - в мобильной версии) в левом верхнем углу блокнота. Чтобы отредактировать созданную ячейку, кликните по ней дважды мышкой и введите описание задачи.

>*Подставляя численные значения для орбиты, расположенной вблизи поверхности Земли (h ≈ 0, M = 5,97·10^24 кг, R0 = 6 371 000 м (радиус указывается в метрах), G=6.67·10^-11 м³·кг⁻¹·с⁻²), получаем*

Для того чтобы создать ячейку с кодом нажмите на кнопку **"+ Код"** (**"+ <>"** - в мобильной версии) в левом верхнем углу блокнота. Чтобы отредактировать созданную ячейку, кликните по ней один раз мышкой и введите код. Для запуска выполнения кода в ячейке, выделите ее и нажмите Ctrl+Enter, либо в меню "Среда выполнения" выберите один из пунктов меню: "Выполнить код в сфокусированной ячейке" или "Запустить выбранный код". Так же можно выполнить весь код последовательно в блокноте, выбрав в меню "Среда выполнения" пункт "Выполнить все".


```python
# Это ячейка с кодом
(6.67*10**(-11) * 5.97 * 10**(24) / 6371000)**(1/2)
```

После выполнения ячейки, результат последней ячейки выводиться в блокнот. Также при установке необходимых библиотек, все предупреждения и информация об установке так же будет выводиться в блокнот. Вы, наверное, заметили, что когда фокусируете мышку на ячейке, то слева от нее появляется значок "треугольник в кружочке", по которому также можно запустить выполнение ячейки с кодом.

Таким образом мы получили, что если запустить камень со скоростью 7900 м/с, то он станет искусственным спутником земли и будет вращаться вокруг земли бесконечно долго, не падая на нее. Данную задачу мы решили с помощью языка программирования **Python**, о котором подробно пойдет речь в следующих уроках.

С помощью символа **#** в Python вставляют текстовый комментарий к коду, делают дополнительное пояснение, документируют код, при этом комментарий не оказывает влияния на работу программы. Начинающему разработчику важно приучить себя комментировать код. Это очень полезный навык, делающий код понятным, читаемым и его легче поддерживать.

**Поддерживать код** - выражение из лексикона разработчика, означает вносить правки в программное обеспечение, исправлять ошибки (как свои так и чужие), обновлять программу, если обновились библиотеки и модули, которые используются программой. Операционные системы тоже обновляются. Поэтому, если код не поддерживать, то со временем программа устареет и перестанет запускаться.
